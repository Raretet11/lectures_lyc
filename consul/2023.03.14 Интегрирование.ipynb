```python
def f(x):  # хотим найти - неизестна нам изначально
    return x**3

def f1(x):  # известна
    return 3*x**2


xlim_min, xlim_max = -11, -8
x = np.arange(xlim_min, xlim_max, 0.1)

y = f(x)

plt.plot(x, y)
# plt.plot(x, f1(x))

x0 = -10   # известно
y0 = -1000   # известно

dx = 1  # мы задаем


def line(x):
    b = y0 - f1(x0)*x0
    return f1(x0) * x + b

plt.plot(x0, y0, '.', color="r")
plt.plot([xlim_min, xlim_max], [line(xlim_min), line(xlim_max)], color="g")

yn = y0
xn = x0

xn1 = xn + dx

plt.hlines(y0, xmin=xn, xmax=xn1, color="r")
plt.vlines(xn1, ymin=y0, ymax=line(xn1), color="r")

# tg(alpha) = y/x  => y = x * tg(alpha)
yn1 = yn + f1(xn) * dx   # line(xn1)

def line1(x):
    b = yn1 - f1(xn1)*xn1
    return f1(xn1) * x + b

plt.plot([xlim_min, xlim_max], [line1(xlim_min), line1(xlim_max)], color="g")
```
